```python
Roosevelt_words = inaugural.words('1941-Roosevelt.txt')
Kennedy_words = inaugural.words('1961-Kennedy.txt')
Nixon_words = inaugural.words('1973-Nixon.txt')
print('Number of words in Speech of President Franklin D. Roosevelt  are', len(Roosevelt_words))
print('Number of words in Speech of President John F. Kennedy  are', len(Kennedy_words))
print('Number of words in Speech of President Richard Nixon  are', len(Nixon_words))
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
all_words_clean = set(stopwords.words('english')+list(string.punctuation))
def clean(array,all_words_clean):
 cleaned = []
 for a in array:
     al = a.lower()
     if al not in all_words_clean and a!='--':
         cleaned.append(al)
 return cleaned
```

```python
Roosevelt_words_clean = clean(Roosevelt_words,all_words_clean)
Kennedy_words_clean = clean(Kennedy_words,all_words_clean)
Nixon_words_clean = clean(Nixon_words,all_words_clean)
```

```python
print('Number of words in Speech of President Franklin D. Roosevelt before removal of stop words is are', len(Roosevelt_words))
print('Number of words in Speech of President John F. Kennedy  before removal of stop words isare', len(Kennedy_words))
print('Number of words in Speech of President Richard Nixon  before removal of stop words isare', len(Nixon_words))
print('Number of words in Speech of President Franklin D. Roosevelt  after removal of stop words is :',len(Roosevelt_words_clean))
print('Number of words in Speech of President John F. Kennedy after removal of stop words is :',len(Kennedy_words_clean))
print('Number of words in Speech of President Richard Nixon after removal of stop words is :',len(Nixon_words_clean))
```

```python
sample_text = """The world is very different now. For man holds in his mortal hands the power to abolish all forms of human poverty and all forms of human life"""
stopwords_clean = set(stopwords.words('english')+list(string.punctuation))
word_tokens = word_tokenize(sample_text)
cleaned_sentence = [w for w in word_tokens if not w.lower() in stopwords_clean]


print("Sample sentence")
print(sample_text)
print("-------------------------")
print("Filtered sentence")
print(cleaned_sentence)
```

```python
print('Words that occured most in speech of  of President Franklin D. Roosevelt after removal of stopwords are ', nltk.FreqDist(Roosevelt_words_clean).most_common(3))
print('Words that occured most in speech of  of President John F. Kennedy after removal of stopwords are ',nltk.FreqDist(Kennedy_words_clean).most_common(3))
print('Words that occured most in speech of  of President Richard Nixon , after removal of stopwords are ',nltk.FreqDist(Nixon_words_clean).most_common(3))
```

```python
from wordcloud import WordCloud
wc_p = ' '.join(Roosevelt_words_clean)
wordcloud = WordCloud(width = 3000, height = 3000,
                background_color ='black',
                min_font_size = 10, random_state=100).generate(wc_p)

# plot the WordCloud image
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.xlabel('Word Cloud')
plt.tight_layout(pad = 0)

print("Word Cloud for President Roosevelt (after cleaning)!!")
plt.show()
```
