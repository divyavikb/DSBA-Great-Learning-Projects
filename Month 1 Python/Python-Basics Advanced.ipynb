## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5

To get the total Revenue we need to first define a function that will compute the Revenue when we give the discounted price.Let call the function revenue(). We will pass the discounted price(dp) as argument and value returned will be the revenue.Using the conditions mentioned above we have a built a nested if elif else condition to compare the discounted prices and return the revenue rev acordingly.
apply() function will apply the function revenue on the column discounted price.
sum() will sum up the revenue
Total_Revenue is a variable that will contain the total Revenue for the E commerce company

```python
def revenue(dp):
    if dp>600:
        rev= dp*0.25
    elif dp>350 and dp<=600:
        rev= dp*0.15
    elif dp>100 and dp<=350:
        rev= dp*0.10
    else:
        rev= dp*0.05
    return rev
```

```python
Total_Revenue =ECom_Data['discounted_price'].apply(revenue).sum()
```

```python
print("Total Revenue generated by the E-Commerce company over all orders placed is ", Total_Revenue,'$',sep='')
```

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days

### Ans 9 a)

Firstly converting the Order _date column which is of Type object to Datetime Type using to_datetime function

```python
ECom_Data['Order_Date']  = pd.to_datetime(ECom_Data['Order_Date'],dayfirst=True)
```

_Creating new column 'month' and extracting month from Order_Date using dt.month

```python
ECom_Data['month'] = ECom_Data['Order_Date'].dt.month
```

_Creating new column 'month' and extracting month name from Order_Date using dt.month_name()

```python
ECom_Data['month_name'] = ECom_Data['Order_Date'].dt.month_name()
```

Creating a new Dataframe Reg_Rev for convinience sake with only Region, Month and Revenue as columns
Firslty group by Region and Month and total the Revenue for that month per Region.Reset Index() gives new index labels for each row

```python
Reg_Rev=pd.DataFrame(ECom_Data.groupby(['Region','month'])['revenue'].sum().reset_index())
```

```python
Reg_Rev
```
