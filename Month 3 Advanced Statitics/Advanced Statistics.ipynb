* the alternative  states that there is a difference in implant hardness among temperature levels for Alloy1, $μ_t1500$ $≠$$μ_t1600$ $≠$$μ_t1700$
* the alternative  states that there is a difference in implant hardness among temperature levels for Alloy2, $μ_t1500$ $≠$$μ_t1600$ $≠$$μ_t1700$

$μ_t1500$ denotes the mean of dental hardness among temperature level of 1700,

### Step 2: Decide the significance level

Here we select $\alpha$ = 0.05 as given in the question.

### Step 3: Identify the test statistic

We do not know the population standard deviation and there are many groups and since we are checking response variable Implant hardness against Temperature,we use the one way ANOVA test.

### Step 4: Calculate the p - value and test statistic

```python
formula = 'Response ~ C(Temp)'
model = ols(formula, dentist_df_Alloy1).fit()
aov_table1 = anova_lm(model)
print(" ANOVA Alloy1", aov_table1)
print('fstat  %1.3f' % aov_table1['F'][0])    
print("p-value:", aov_table1['PR(>F)'][0])
formula = 'Response ~ C(Temp)'
model = ols(formula, dentist_df_Alloy2).fit()
aov_table2 = anova_lm(model)
print("ANOVA Alloy2",aov_table2)
print('fstat  %1.3f' % aov_table2['F'][0])    
print("p-value:", aov_table2['PR(>F)'][0])
```

### Step 5:  Decide to reject or accept null hypothesis

```python
# p_value < 0.05 => alternative hypothesis:
# they don't have the same mean at the 5% significance level


alpha_level = 0.05

if (aov_table1['PR(>F)'][0]) < alpha_level:
    print('We have enough evidence to reject the null hypothesis in favour of alternative hypothesis for Alloy 1.')
    print('We conclude that Means of Implant Hardness is not same across all Temperatures for Alloy 1 .There is a difference in atleast one pair')
else:
    print('We do not have enough evidence to reject the null hypothesis in favour of alternative hypothesis for Alloy 1.')
    print('We conclude that Means of Implant Hardness is same across all Temperatures for Alloy 1')
  
    
if (aov_table2['PR(>F)'][0]) < alpha_level:
    print('We have enough evidence to reject the null hypothesis in favour of alternative hypothesis for Alloy 2.')
    print('We conclude that Means of Implant Hardness is not same across all Temperatures for Alloy 2.There is a difference in atleast one pair')
else:
    print('We do not have enough evidence to reject the null hypothesis in favour of alternative hypothesis for Alloy 2.')
    print('We conclude that Means of Implant Hardness is same across all Temperatures for Alloy 2')
  
```
